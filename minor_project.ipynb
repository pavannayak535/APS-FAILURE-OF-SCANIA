```python
pca.n_components = 170
pca_data = pca.fit_transform(sample_data)

percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_);

cum_var_explained = np.cumsum(percentage_var_explained)

# Plot the PCA spectrum
plt.figure(1, figsize=(8, 8))
sns.set_style("whitegrid");
plt.clf()
plt.plot(cum_var_explained, linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n_components')
plt.ylabel('Cumulative_explained_variance')
plt.show()
```

# IMPORTING REQUIRED LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
import xgboost as xgb
from tqdm import tqdm
from sklearn.metrics import f1_score, log_loss, confusion_matrix,classification_report
import scikitplot as skplt
from sklearn.metrics import accuracy_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler
```

F1 SCORE && LOG-LOSS

```python
def evaluate(y_test,y_pred,y_pred_proba):
    if len(y_pred)>0:
        f1 = f1_score(y_test,y_pred,average="weighted")
        print("F1 score: ",f1)
    if len(y_pred_proba)>0:
        logloss = log_loss(y_test,y_pred_proba, eps=1e-15, normalize=True, sample_weight=None, labels=None)
        print("Log loss for predicted probabilities:",logloss)
```

```python
y_train = aps['class'].copy(deep=True)
X_train = d.copy(deep=True)
y_test = data_test['class'].copy(deep=True)
X_test = data_test.copy(deep=True)
X_test.drop(['class'], inplace=True, axis=1)
```

# LOGISTIC REGRESSION

```python
 from sklearn.linear_model import LogisticRegression
logistic_clf= LogisticRegression(random_state=0, solver='lbfgs',C=1.0).fit(X_train,y_train)
y_pred_lr= logistic_clf.predict(X_test)
y_pred_proba_lr = logistic_clf.predict_proba(X_test)
evaluate(y_test,y_pred_lr,y_pred_proba_lr)
```

```python
import scikitplot as skplt
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_lr ).ravel()
skplt.metrics.plot_confusion_matrix(y_test, y_pred_lr, normalize=False)
plt.show()
print(classification_report(y_test,y_pred_lr))
```

```python
from sklearn.metrics import auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred_lr)
roc_auc = auc(fpr, tpr)


plt.figure()
plt.plot(fpr, tpr, color='red', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
y_test_predictions_prec = y_pred_proba_lr[:,1] > 0.85
```

```python
skplt.metrics.plot_confusion_matrix(y_test, y_test_predictions_prec, normalize=False)
plt.show()
print(classification_report(y_test, y_test_predictions_prec))
```

```python
scores = logistic_clf.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, scores)
```

```python
min_cost = np.inf
best_threshold = 0.5
costs = []
for threshold in tqdm(thresholds):
    y_pred_threshold = scores > threshold
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_threshold).ravel()
    cost = 10*fp + 500*fn
    costs.append(cost)
    if cost < min_cost:
        min_cost = cost
        best_threshold = threshold
print("Best threshold: {:.4f}".format(best_threshold))
print("Min cost: {:.2f}".format(min_cost))
```

```python
y_pred_test_lr = logistic_clf.predict_proba(X_test)[:,1] > best_threshold
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test_lr).ravel()
skplt.metrics.plot_confusion_matrix(y_test,y_pred_test_lr, normalize=False)
10*fp + 500*fn
```
